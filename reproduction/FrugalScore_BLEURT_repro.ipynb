```python
import datasets
import numpy as np

from geneval.geneval.data.wmt import WMT18
```

```python
# kendall score
# https://github.com/Tiiiger/bert_score/blob/master/reproduce/get_wmt18_seg_results.py#L53

def kendall_score(scores_better, scores_worse):
    total = len(scores_better)
    correct = np.sum(np.array(scores_better) > np.array(scores_worse))
    incorrect = total - correct
    return (correct - incorrect)/total
```

### FrugalScore

https://huggingface.co/spaces/evaluate-metric/frugalscore

```python
frugal_score = datasets.load_metric("frugalscore")
```

```python
frugal_score_results = {}
for lang_pair in ["cs-en", "de-en", "et-en", "fi-en", "ru-en", "tr-en", "zh-en"]:
    # load data
    wmt = WMT18(lang_pair, root="/tmp")
    # compute scores
    scores_better = frugal_score.compute(predictions = wmt.translations_better,
        references = wmt.references
    )
    scores_worse = frugal_score.compute(
        predictions = wmt.translations_worse,
        references = wmt.references
    )
    ks = kendall_score(scores_better["scores"], scores_worse["scores"])
    frugal_score_results[lang_pair] = ks
```

```python
bleurt = datasets.load_metric("bleurt")
```

```python
predictions = ["hello there", "general kenobi"]
references = ["hello there", "general kenobi"]
results = bleurt.compute(predictions=predictions, references=references)

print(results)
```

```python
bleurt_results = {}
for lang_pair in ["cs-en", "de-en", "et-en", "fi-en", "ru-en", "tr-en", "zh-en"]:
    # load data
    wmt = WMT18(lang_pair, root="/tmp")
    # compute scores
    scores_better = bleurt.compute(
        predictions = [wmt.translations_better],
        references = [wmt.references]
    )
    scores_worse = bleurt.compute(
        predictions = [wmt.translations_worse],
        references = [wmt.references]
    )
    ks = kendall_score(scores_better["scores"], scores_worse["scores"])
    bleurt_results[lang_pair] = ks
```
